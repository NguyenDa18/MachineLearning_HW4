```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import decomposition

np.random.seed(91)

X = np.random.rand(200)

X = np.random.randn(200) + 2*X 
Y = np.random.randn(200) + 2*X
Z = np.random.randn(200) + 2*X

data = np.array((X,Y,Z)).T

```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X,Y,Z, c='r')
plt.pause(5)

pca_model = decomposition.PCA(n_components=2)
variance = pca_model.fit(data)
print(variance)
print(pca_model.explained_variance_)
print(pca_model.explained_variance_ratio_)
print(pca_model.explained_variance_ratio_.cumsum() )

pca_data = pca_model.transform(data)

# ax.scatter(pca_data[:,0], pca_data[:,1], pca_data[:,2], c='b')

# plt.show()
```
